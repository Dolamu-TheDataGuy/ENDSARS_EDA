# Importing Data

```r

endsars2021 <- read.csv("endsars_2021.csv")

head(endsars2021,3)
```

# Text Cleaning

```r
#working on endsars2021
corpus2021<- iconv(endsars2021$content)
#remove retweets
corpus2021<- gsub("(RT|via)((?:\\b\\w*@\\w+)+)"," ",corpus2021)
#remove people
corpus2021<- gsub("@\\w+"," ",corpus2021)
corpus2021<- gsub("\n.*"," ",corpus2021)

######creating a corpus###############
corpus2021<- Corpus(VectorSource(corpus2021))
corpus2021<- tm_map(corpus2021,PlainTextDocument)
# inspect(corpus2021[1:5])
################cleaning text#####################
corpus2021<-tm_map(corpus2021,tolower)
#inspect(corpus2021[1:5])
corpus2021<-tm_map(corpus2021,removePunctuation)
corpus2021<-tm_map(corpus2021,removeNumbers)

cleanset2021<- tm_map(corpus2021, removeWords,stopwords('english'))
cleanset2021<- tm_map(cleanset2021,stemDocument)
#inspect(cleanset2021[1:5])
removeURL2021 <- function(x) gsub('http[[:alnum:]]*','',x)
cleanset2021<- tm_map(cleanset2021,content_transformer(removeURL2021))
cleanset2021<- tm_map(cleanset2021,removeWords,c('rt'))
#cleanset<- tm_map(cleanset, gsub,
 #                 pattern ='…',
  #                replace = '')
inspect(cleanset2021[1:5])
```

```r
###########Term document matrix##########
tdm2021<- TermDocumentMatrix(cleanset2021)
# tdm2021
tdm2021<- as.matrix(tdm2021)
# tdm2021<- as.matrix(tdm2021)

freq2021<-rowSums(tdm2021)
# freq2021

freq2021<-subset(freq2021,freq2021>=600)
```

# Frequent Word Count

```r
#########Bar plot########


png("word_count2021.png", width = 700, height = 350)

barplot(freq2021,
        las = 2,
        col = rainbow(30), ylab = 'tweet frequency')

dev.off()
```
