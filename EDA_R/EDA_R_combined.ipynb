# Importing Data

```r
all_data <- read.csv("combined_scraped_data.csv")


head(all_data,3)
```

# Text Cleaning

```r
#working on endsars20all
corpus20all<- iconv(all_data$content)
#remove retweets
corpus20all<- gsub("(RT|via)((?:\\b\\w*@\\w+)+)"," ",corpus20all)
#remove people
corpus20all<- gsub("@\\w+"," ",corpus20all)
corpus20all<- gsub("\n.*"," ",corpus20all)

######creating a corpus###############
corpus20all<- Corpus(VectorSource(corpus20all))
corpus20all<- tm_map(corpus20all,PlainTextDocument)
# inspect(corpus20all[1:5])
################cleaning text#####################
corpus20all<-tm_map(corpus20all,tolower)
#inspect(corpus20all[1:5])
corpus20all<-tm_map(corpus20all,removePunctuation)
corpus20all<-tm_map(corpus20all,removeNumbers)

cleanset20all<- tm_map(corpus20all, removeWords,stopwords('english'))
cleanset20all<- tm_map(cleanset20all,stemDocument)
#inspect(cleanset20all[1:5])
removeURL20all <- function(x) gsub('http[[:alnum:]]*','',x)
cleanset20all<- tm_map(cleanset20all,content_transformer(removeURL20all))
cleanset20all<- tm_map(cleanset20all,removeWords,c('rt'))
#cleanset<- tm_map(cleanset, gsub,
 #                 pattern ='…',
  #                replace = '')
inspect(cleanset20all[1:5])
```

```r
###########Term document matrix##########
tdm20all<- TermDocumentMatrix(cleanset20all)
# tdm20all
tdm20all<- as.matrix(tdm20all)
# tdm20all<- as.matrix(tdm20all)

freq20all<-rowSums(tdm20all)
# freq20all

freq20all<-subset(freq20all,freq20all>=600)
```

# Frequent Word Count

```r
#########Bar plot########


png("word_count20all.png", width = 650, height = 350)

barplot(freq20all,
        las = 2,
        col = rainbow(30), ylab = 'tweet frequency')

dev.off()
```
