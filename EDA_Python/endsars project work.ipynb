```r
#working on endsars23
corpus23<- iconv(endsars23$text)
#remove retweets
corpus23<- gsub("(RT|via)((?:\\b\\w*@\\w+)+)"," ",corpus23)
#remove people
corpus23<- gsub("@\\w+"," ",corpus23)
corpus23<- gsub("\n.*"," ",corpus23)
```

```r
######creating a corpus###############
corpus23<- Corpus(VectorSource(corpus23))
corpus23<- tm_map(corpus23,PlainTextDocument)
inspect(corpus23[1:5])
```

```r
################cleaning text#####################
corpus23<-tm_map(corpus23,tolower)
#inspect(corpus23[1:5])
corpus23<-tm_map(corpus23,removePunctuation)
corpus23<-tm_map(corpus23,removeNumbers)
```

```r
cleanset23<- tm_map(corpus23, removeWords,stopwords('english'))
cleanset23<- tm_map(cleanset23,stemDocument)
#inspect(cleanset23[1:5])
removeURL23 <- function(x) gsub('http[[:alnum:]]*','',x)
cleanset23<- tm_map(cleanset23,content_transformer(removeURL23))
cleanset23<- tm_map(cleanset23,removeWords,c('rt'))
```
